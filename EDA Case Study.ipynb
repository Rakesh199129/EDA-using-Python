# Data Cleaning

```python
# looking for the columns with high null values

round(100*(loan.isnull().sum()/len(loan.index)),2)
```

```python
# deleting those columns with 100% missing values

loan = loan.drop(['total_il_high_credit_limit','total_bc_limit','total_bal_ex_mort','tot_hi_cred_lim',
                 'mo_sin_old_rev_tl_op','mo_sin_rcnt_rev_tl_op','mo_sin_rcnt_tl','mort_acc','mths_since_recent_bc',
                  'mths_since_recent_bc_dlq','mths_since_recent_inq','mths_since_recent_revol_delinq','num_accts_ever_120_pd',
                  'num_actv_bc_tl','num_actv_rev_tl','num_bc_sats','num_bc_tl','num_il_tl','num_op_rev_tl','num_rev_accts',
                  'num_rev_tl_bal_gt_0','num_sats','num_tl_120dpd_2m','num_tl_30dpd','num_tl_90g_dpd_24m','num_tl_op_past_12m',
                  'pct_tl_nvr_dlq','percent_bc_gt_75'], axis=1)
loan.head(5)
```

```python
# repeating the same step to check for columns with high null values

round(100*(loan.isnull().sum()/len(loan.index)),2)
```

```python
# dropping those columns with 100% missing values

loan = loan.drop(['annual_inc_joint','dti_joint','verification_status_joint','tot_coll_amt','tot_cur_bal','open_acc_6m',
                  'open_il_6m','open_il_12m','open_il_24m','mths_since_rcnt_il','total_bal_il','il_util','open_rv_12m',
                  'open_rv_24m','max_bal_bc','all_util','total_rev_hi_lim','inq_fi','total_cu_tl','inq_last_12m',
                  'acc_open_past_24mths','avg_cur_bal','bc_open_to_buy','bc_util'], axis=1)
loan.head(5)
```

```python
# repeating the same step to check for high null values

round(100*(loan.isnull().sum()/len(loan.index)),2)
```

```python
# dropping the columns with high null values

loan = loan.drop(['mo_sin_old_il_acct','mths_since_last_major_derog','next_pymnt_d','mths_since_last_record'], axis=1)
loan
```

### General inspection shows we may not need few other columns like member id, id, emp title, url, desc, title, zip code, etc., for data analysis. so, we would delete these columns from the data set.

```python
loan = loan.drop(['id','member_id','emp_title','url','desc','title','zip_code'],axis=1)
loan.head(5)
```

```python
# looking at the application type

loan['application_type'].value_counts()
```

### So, the given data set is only of INDIVIDUAL application type. As this is understood, we can drop this column. Similarly, columns like chargeoff_within_12_mths, delinq_amnt, pub_rec_bankruptcies, tax_liens, collections_12_mths_ex_med, can also be deleted, as they have maximum of zero entries.

```python
# dropping un necessary columns

loan = loan.drop(['application_type','chargeoff_within_12_mths','delinq_amnt','tax_liens',
                 'collections_12_mths_ex_med'],axis = 1)
loan
```

#### Let us convert the int_rate column from string type to float type which might be useful for our analysis.

```python
# converting the int_rate column to float type

loan['int_rate'] = loan['int_rate'].str[:-1].astype('float64')
loan['int_rate']
```

#### Let us look at the issue date
The issue date column has to be converted to the standard date-time format first..

```python
# looking at the issue_d column of the given dataset

loan['issue_d']
```

```python
# checking for null values in the issue_d column

loan['issue_d'].isnull().sum()
```

We are creating two new columns here called issue_yr (for year) and issue_mon (for month)from the column issue_d by splitting it. Here issue_yr has only two digit value of year, we can replace it by the usual 4 digit year notation in the next step.

```python
# splitting the issue_d column into two new columns for month and year

loan['issue_mon'] = loan.issue_d.str.split('-', expand = True)[0]
loan['issue_yr'] = loan.issue_d.str.split('-', expand = True)[1]
loan
```

```python
# replacing entries in issue_yr column with ususal 4 digit notation

loan['issue_yr'] = loan.issue_yr.replace("11","2011")
loan['issue_yr'] = loan.issue_yr.replace("10","2010")
loan['issue_yr'] = loan.issue_yr.replace("09","2009")
loan['issue_yr'] = loan.issue_yr.replace("08","2008")
loan['issue_yr'] = loan.issue_yr.replace("07","2007")
loan['issue_yr']
```

### Let us now analyse on Loan status column

```python
# we now try to use subplots to plot the data based on loan status

fig = plt.figure(figsize=(16,12))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

loans_addr_state = loan.groupby(['grade', 'loan_status']).size()
loans_addr_state.unstack().plot(kind='bar', stacked=True, ax=ax1, grid=False)
ax1.set_title('Type of Loans by Grade', fontsize=14)


loans_by_grade = loan.groupby(['sub_grade', 'loan_status']).size()
loans_by_grade.unstack().plot(kind='bar', stacked=True, ax=ax2, grid=False)
ax2.set_title('Type of Loans by Sub-Grade', fontsize=14)

by_interest = loan.groupby(['issue_yr', 'loan_status']).int_rate.mean()
by_interest.unstack().plot(ax=ax3)
ax3.set_title('Average Interest rate by loan status', fontsize=14)
ax3.set_ylabel('Interest Rate (%)', fontsize=12)
```

So here we observe that the loans whose interest rate is high are the ones with high charged off loan status.

### Creating new column "balance" which is the difference between total payment and funded amount

```python
# defining a new column called balance
loan['balance']=loan['total_pymnt'].astype('float64') - loan['funded_amnt'].astype('float64')
loan['balance']
```

```python
# defining new column profit and loss where negative values imply loss and positive imply profit percentages
loan['profit&loss']=loan['balance']/(loan['funded_amnt'].astype('float64'))*100
loan.head()
```

```python
# let us now try to plot these values
plt.figure(figsize=(8,8))

sns.barplot(x='loan_status',y='profit&loss',data=loan)
plt.xlabel("Loan Status")
plt.ylabel("Profit and Loss")

plt.title("Loan Status with Profit and loss")

plt.show()
```

#### So as the plot shows the loan status "charged off" has negative values indicating loss to the company..

```python
plt.figure(figsize=(11,11))

sns.barplot(x='term',y='profit&loss',data=loan,hue='grade')
plt.xlabel("grade")
plt.ylabel("Profit and Loss")
```

#### This shows clearly that customers under grade G category with 36 month loan term period tend to show loss

```python
#Funded amount with Terms
# Relationship of default with term with a barplot

plt.figure(figsize=(8,15))
sns.barplot(y='emp_length',x='profit&loss',data=loan,hue='grade')
plt.ylabel("Emp Length")
plt.xlabel("Profit and Loss")

plt.title("Employee work Length vs Profit/Loss and Grade")
```
